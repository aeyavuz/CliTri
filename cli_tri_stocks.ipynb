```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
import seaborn as sns
import os
from xml.etree import ElementTree
import datetime
import warnings
warnings.filterwarnings('ignore')
```

```python
#title

df_trials = pd.DataFrame()
df = pd.DataFrame()
i=0
list_keywords = []


# Read in data from all folders/subfolders
data_path = './AllPublicXML'

for temp in os.listdir(data_path)[0:]:
    if (temp != 'AllPublicXML') & (temp != 'Contents.txt') & (temp[0] != '.'):
        temp_path = os.path.join(data_path, temp)
        print("processing: ", temp_path)
        for filename in os.listdir(temp_path):
            filepath = os.path.join(temp_path, filename)
            #print(filepath)

            tree = ElementTree.parse(filepath)
            root = tree.getroot()  #read xml root

            trial = {}
            trial['id'] = root.find('id_info').find('nct_id').text
            trial['overall_status'] = root.find('overall_status').text
            trial['study_type'] = root.find('study_type').text
            
            if root.find('start_date') != None:
                trial['start_date'] = root.find('start_date').text
            else:
                 trial['start_date'] = ''

                    
            if root.find('completion_date') != None:
                trial['completion_date'] = root.find('completion_date').text
            else:
                 trial['completion_date'] = ''

            if root.find('study_first_submitted') != None:
                trial['study_first_submitted'] = root.find('study_first_submitted').text
            else:
                 trial['study_first_submitted'] = ''                    
                    
            if root.find('enrollment') != None:
                trial['enrollment'] = root.find('enrollment').text
            else:
                 trial['enrollment'] = ''
                    
            if root.find('condition') != None:
                trial['condition'] = root.find('condition').text
            else:
                trial['condition'] = ''
            
            if root.find('location_countries') != None:
                trial['location_countries'] = root.find('location_countries').find('country').text
            else:
                trial['location_countries'] = ''

            if root.find('intervention') != None:
                trial['intervention'] = root.find('intervention').find('intervention_name').text
            else:
                trial['intervention'] = ''
            
            if root.find('phase') != None:
                trial['phase'] = root.find('phase').text
            else:
                trial['phase'] = ''
                
            if root.find('overall_status') != None:
                trial['overall_status'] = root.find('overall_status').text
            else:
                trial['overall_status'] = ''                

            if root.find('intervention') != None:
                trial['intervention_type'] = root.find('intervention').find('intervention_type').text
            else:
                trial['intervention_type'] = ''
                
                
            for entry in root.findall('keyword'):
                list_keywords.append(entry.text)
                if root.find('official_title') == None:
                    trial['title'] = root.find('brief_title').text
                else:
                    trial['title'] = root.find('official_title').text
                    
            date_string = root.find('required_header').find('download_date').text
            trial['date_processed'] = date_string.replace('ClinicalTrials.gov processed this data on ', '')

            trial['sponsors'] = root.find('sponsors').find('lead_sponsor').find('agency').text
            
            if root.find('sponsors').find('lead_sponsor').find('agency_class') != None:
                trial['sponsors_type'] = root.find('sponsors').find('lead_sponsor').find('agency_class').text
            else:
                trial['sponsors_type'] = ''
            
#             if root.find('brief_summary').find('textblock') != None:
#                 trial['summary'] = root.find('brief_summary').find('textblock')
#             else:
#                 trial['summary'] = ''
                

            df  = pd.DataFrame(trial,index=[i])
            i=i+1


            df_trials = pd.concat([df_trials, df])
```
